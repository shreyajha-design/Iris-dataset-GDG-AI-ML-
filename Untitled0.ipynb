```python
colors=['red','yellow','blue']
species=['setosa','versicolor','virginica']
```

```python
for i in range(3):
  x=df[df['species']==species[i]]
  py.scatter(x['sepal_length'],x['sepal_width'],c=colors[i],label=species[i])
py.xlabel("Sepal Length")
py.ylabel("Sepal Width")
py.legend()
```

```python
for i in range(3):
  x=df[df['species']==species[i]]
  py.scatter(x['petal_length'],x['petal_width'],c=colors[i],label=species[i])
py.xlabel("Petal Length")
py.ylabel("Petal Width")
py.legend()
```

```python
for i in range(3):
  x=df[df['species']==species[i]]
  py.scatter(x['sepal_length'],x['petal_length'],c=colors[i],label=species[i])
py.xlabel("Sepal Length")
py.ylabel("Petal Length")
py.legend()
```

```python
for i in range(3):
  x=df[df['species']==species[i]]
  py.scatter(x['sepal_width'],x['petal_width'],c=colors[i],label=species[i])
py.xlabel("Sepal Width")
py.ylabel("Petal Width")
py.legend()
```

```python
flower_mapping={'setosa':0,'versicolor':1, 'virginica':2}
df["species"]=df["species"].map(flower_mapping)
df.head()
```

```python
X=df[['sepal_length','sepal_width','petal_length','petal_width']].values
y=df[['species']].values
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
expected=y
predicted=model.predict(X)
predicted
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```
